```python
import os
import pandas as pd

# Read the Excel data
data = pd.read_excel(r"C:\Users\ravit\Downloads\data day4\day4.xlsx")

# Get the photo data column name
photo_data_column_name = "no"

# Get the category data column name
category_data_column_name = "f.r"

# Create three empty lists to store the photos in each category
over_ripe_photos = []
early_ripe_photos = []
pre_ripe_photos = []

# Loop through the Excel data
for index, row in data.iterrows():
    # Get the photo filename
    photo_filename = row[photo_data_column_name] + ".jpg"

    # Get the photo category
    photo_category = row[category_data_column_name]

    # Check if the photo file exists
    if os.path.exists(r"C:\Users\ravit\Downloads\data day4\FLUKE/" + photo_filename):
        # Add the photo to the appropriate list
        if photo_category == "Over Ripe":
            over_ripe_photos.append(photo_filename)
        elif photo_category == "Early Ripe":
            early_ripe_photos.append(photo_filename)
        else:
            pre_ripe_photos.append(photo_filename)
    else:
        print(f"File not found: {photo_filename}")

# Create three directories to store the sorted photos
os.makedirs(r"C:\Users\ravit\Downloads\data day4\sorted_photos\Over Ripe", exist_ok=True)
os.makedirs(r"C:\Users\ravit\Downloads\data day4\sorted_photos\Early Ripe", exist_ok=True)
os.makedirs(r"C:\Users\ravit\Downloads\data day4\sorted_photos\Pre Ripe", exist_ok=True)

# Move the sorted photos to the appropriate directories
for photo_filename in pre_ripe_photos:
    source_path = r"C:\Users\ravit\Downloads\data day4\FLUKE" + "/" + photo_filename
    destination_path = r"C:\Users\ravit\Downloads\data day4\sorted_photos\Pre Ripe" + "/" + photo_filename
    os.rename(source_path, destination_path)

for photo_filename in early_ripe_photos:
    source_path = r"C:\Users\ravit\Downloads\data day4\FLUKE" + "/" + photo_filename
    destination_path = r"C:\Users\ravit\Downloads\data day4\sorted_photos\Early Ripe" + "/" + photo_filename
    os.rename(source_path, destination_path)

for photo_filename in over_ripe_photos:
    source_path = r"C:\Users\ravit\Downloads\data day4\FLUKE" + "/" + photo_filename
    destination_path = r"C:\Users\ravit\Downloads\data day4\sorted_photos\Over Ripe" + "/" + photo_filename
    os.rename(source_path, destination_path)

print("The photos have been sorted into three categories.")
```
